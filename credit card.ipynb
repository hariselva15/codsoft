```python
data=data.drop('category',axis=1)
data
```

```python
legit = data[data.is_fraud == 0]
fraud = data[data.is_fraud == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.merch_long.describe()
```

```python
fraud.merch_long.describe()
```

```python
legit_sample = legit.sample(n=85)
legit_sample
```

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['is_fraud'].value_counts()
```

```python
#new_dataset.groupby('is_fraud').mean()
```

```python
x = new_dataset.drop(columns='is_fraud',axis=1)
y = new_dataset['is_fraud']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test ,y_train,y_test=train_test_split(x , y , test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python

# logistic regression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,  y_train)
```

```python
print("accuracy_on_training_data: ",training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python

print('accracy score on test data :',test_data_accuracy)
```

```python

import joblib
```

```python
filename = 'finalized_model.save'
joblib.dump(model, filename)
```

```python

loaded_model = joblib.load(filename)
result = loaded_model.score(x_test, y_test)
print(result)
```
